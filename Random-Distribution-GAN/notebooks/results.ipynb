# Results of the project

```python
import os

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.set_style('whitegrid')
```

## Training data

```python
normal_sample = np.loadtxt(os.path.join('..', 'data', 'normal_sample.txt'))
uniform_sample = np.loadtxt(os.path.join('..', 'data', 'uniform_sample.txt'))
exponential_sample = np.loadtxt(os.path.join('..', 'data', 'exponential_sample.txt'))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

sns.histplot(normal_sample, kde=True, ax=ax[0])
sns.histplot(uniform_sample, kde=True, ax=ax[1])
sns.histplot(exponential_sample, kde=True, ax=ax[2])

for i in range(3):
    ax[i].set_xlabel('x')
    ax[i].set_ylabel('p(x)')
    ax[i].set_title(f'Histogram of {["normal", "uniform", "exponential"][i]} sample')

plt.show()
```
