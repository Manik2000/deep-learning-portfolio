```python
import os
import warnings

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.set(style="whitegrid")
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
df = pd.read_csv(os.path.join("..", "..", "data", "raw", "MSFT.csv"))
```

```python
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.isna().sum()
```

```python
melted_df = pd.melt(df.loc[:, df.columns != "Volume"], id_vars=["Date"], var_name="Price Type", value_name="Price")

sns.lineplot(x="Date", y="Price", hue="Price Type", data=melted_df)
plt.xticks(rotation=45)
plt.title("Microsoft Stock Prices")
plt.show()
```
